```python
#Add Outlier Detection using IQR
Q1 = df["Popularity"].quantile(0.25)
Q3 = df["Popularity"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + (1.5 * IQR)
outliers = df[df["Popularity"]> upper_bound]
outliers[["Title", "Popularity"]].sort_values(by ="Popularity", ascending = False)
```

```python
# Skewness Analysis
popularity_skew = df['Popularity'].skew()
vote_count_skew = df['Vote_Count'].skew()
pd.DataFrame({
    'Metric': ['Popularity', 'Vote_Count'],
    'Skewness_Score': [popularity_skew, vote_count_skew]
})
```

```python
# Adding Log Transformation for Popularity
df["Popularity_Log"] = np.log(df["Popularity"])
new_skew = df["Popularity_Log"].skew()
pd.DataFrame({
    "Metric":["Original Popularity", "Log Popularity"],
    "Skewness_Score":[df["Popularity"].skew(), new_skew]
})
```

```python
#Comparing Original vs Log Histogram
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.histplot(df["Popularity"], bins=30, kde=True)
plt.title("Before: Original Popularity (High Skewness)")
plt.subplot(1, 2, 2)
sns.histplot(df["Popularity_Log"], bins=30, kde=True, color="salmon")
plt.title("After: Log Popularity (Balanced)")
plt.show()
```

```python
#Creating Success Score
min_Popularity = df["Popularity_Log"].min()
max_popularity = df["Popularity_Log"].max()
df["normal_Popularity"] = (df["Popularity_Log"] - min_Popularity) / (max_popularity - min_Popularity)
max_Rating = df["Vote_Average"].max()
min_Rating = df["Vote_Average"].min()
df["normal_Rating"] = (df["Vote_Average"] - min_Rating) / (max_Rating - min_Rating)
df["success_score"] = ((df["normal_Popularity"]*0.5)+(df["normal_Rating"]*0.5))*100
df[["Title", "success_score"]]
```

### Observation
The custom success score provides a balanced evaluation by combining multiple performance indicators. Movies with higher success scores generally demonstrate strong ratings, popularity, and audience engagement simultaneously.

```python
#Creating Rating Categories
def get_label(Vote_Average):
    if Vote_Average >= 8.0:
        return "Masterpiece"
    elif Vote_Average >= 7.0:
        return "Hit"
    elif Vote_Average >= 5.0:
        return "Average"
    else:
        return "Flop"

df["Rating_category"] = df["Vote_Average"].apply(get_label)
df[["Title", "Vote_Average", "Rating_category", "success_score"]]
```

```python
#Creating Popularity Categories
def get_pop_label(Popularity):
    if Popularity >= 500:
        return "Global Blockbuster"
    elif Popularity >= 100:
        return "Trending Hit"
    elif Popularity >= 30:
        return "Mainstream"
    else:
        return "Indie/Niche"

df["Popularity_Category"] = df["Popularity"].apply(get_pop_label)
df[["Title", "Vote_Average", "Rating_category", "Popularity_Category", "success_score"]]
```

# Final Conclusion

In the current work, an Exploratory Data Analysis (EDA) was performed on the Netflix Movies Dataset to investigate patterns related to the genre types, ratings, popularity, audience engagement, and performance of movies. The process involved the use of data preprocessing methods to prepare the dataset, after which visualization and statistical techniques were applied to find the patterns. Apart from analyzing the movies that had a high performance, hidden gems of movies, audience rating distribution, language performance of the movies, performance depending on release years, and even defining a custom success score for each movie, the current research identified certain patterns.
